```python
from math import sin, cos, tan, pi
from pylab import *
```

```python
def calcula_f(x):
 return sin(x)
```

```python
x = arange(-2*pi,2*pi,0.01)
```

```python
y = calcula_f(x)
```

```python
plot(x,y)
xlabel("Variavel Independente")
ylabel("Função y = sin(x)")
title("Função seno")
grid(True)
axis([-3*pi, 3*pi,-2,2])
axhline(y = 0, color = 'k')
axvline(x = 0,color = 'k')
```
