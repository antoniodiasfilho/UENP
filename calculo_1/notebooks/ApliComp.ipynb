Numerador e Denominador

```python
x = symbols("x")
expr1 = (x**2+1) / x
expr1
```

```python
num, den = fraction(expr1)
display(num,den)
```

```python
num = numer(expr1)
den = denom(expr1)
display(num,den)
```

```python
from sympy import solve, solveset, symbols, Symbol, sin, Eq, Abs, Reals, FiniteSet, cos, Intersection
```

```python
p, V, R, T = symbols("p, V, R, T")

eq = p * V - R*T

s1 = solve(eq, T)

s2 = solveset(eq, T)

display(eq, s1, s2)
```

```python
print(type(s2))
```

```python
print(type(s1))
```

```python
x = Symbol("x")

eq = Eq(sin(x),1)

s1 = solve(eq)

s2 = solveset(eq)

display(eq,s1,s2)
```

```python
print(s2.func)
s2.args
```

```python
s2.lamda(1)
```

```python
x = Symbol("x",real = True)

eq = Eq(Abs(sin(x)),1)

s1 = solve(eq)

display(eq, s1)
```

```python
i = sin(x) + cos(x) >= 0

s1 = solve(i,x)

s2 = solveset(i,x, domain = Reals)

display(i, s1, s2)
```
