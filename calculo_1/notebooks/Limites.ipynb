```python
from pylab import *
```

```python
def calcula_f(x):
  return (x-1)/(x**2-1)
```

```python
x = arange(0.1,2,0.01)
```

```python
y = calcula_f(x)
```

```python
plot(x,y)
xlabel('Variavel Independente')
ylabel('Função y = (x-1)/(x^2 - 1)')
title('Visão Geometrica do conceito de Limite')
grid(True)
axis([-1,2,0,1])
axhline(y=0, color = 'k')
axvline(x=0, color = 'k')
```
