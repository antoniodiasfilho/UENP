FORMATAÇÃO DOS EIXOS

```python
from pylab import xlabel, ylabel, title, grid, axhline, axvline
```

```python
x = arange(-2,4,0.5)
y = 3*x
```

```python
plot(x,y)

xlabel('Variavel independente x')

ylabel('Função y = 3x')

grid(True)

axhline(y = 0,color = 'k')

axvline(x = 0, color = 'k')

title('GRAFICO DA FUNÇÃO y = 3x')
```
